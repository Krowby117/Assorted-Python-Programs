#### 1. Two factorial functions. First is unoptimized while the second is optimized.

```python
def func(n: int) -> int:
    if (n < 2):
        return 1
    else:
        return func(n-1) * n
    
print(func(1000))
```

```python
def helper(n, result):
    if (n == 0):
        return result
    else:
        return helper(n-1, n * result)
    
def factorial(n):
    return helper(n,1)

print(factorial(1000))
```

#### 2. Towers of Hanoi function that retruns a list of all the moves taken.

```python
def hanoi_moves(n: int, A: str, B: str, C: str)->list:
    if n == 1: 
        return [(A, B)]
    return hanoi_moves(n-1, A, C, B) + [(A, B)] + hanoi_moves(n-1, C, B, A)
    
print(hanoi_moves(3, "A", "B", "C"))
```

#### 3. The Lattice Problem. One with down and right and one with added diagonal move

```python
def count_moves(i: int, j:int, R:int, C:int)->int:
    if i > R or j > C: return 0
    elif i == R and j == C: return 1
    else: return count_moves(i+1, j, R, C) + count_moves(i, j+1, R, C)

print(count_moves(0,0,2,2))
```

#### 4. Max Palindrone function that just finds the max palindrone in a given string

```python
def maxPalindrone(s):
    size = len(s)
    table = [[False] * size for _ in range(size)]

    start = 0
    maxL = 1

    # all substrings of 1 are true
    for i in range(size):
        table[i][i] = True

    # check all substrings of 2
    for i in range(size-1):
        if (s[i] == s[i+1]):
            table[i][i+1] = True
            if maxL == 1:
                start = i
                maxL = 2

    # checl all substrings of 3+
    for length in range(3, size+1):
        for i in range(size - length + 1):
            j = i + length - 1
            if s[i]==s[j] and table[i+1][j-1]:
                table[i][j] = True
                if length > maxL:
                    start = i
                    maxL = length
    
    return s[start:start + maxL]

string = "babad"
print(f"Original string: {string}")
print(f"Max Palindrone : {maxPalindrone(string)}")
```

#### 5. Reverse number function that reverses the digits in a given number

```python
def reverseNum(x):
    temp = ""
    sign = 1

    if x < 0: 
        sign = -1
        x = x * -1

    n = len(str(x))

    for i in range(n):
        temp += str((x // 10**(i)) % 10)

    temp = int(temp) * sign
    return temp

num = -123
print(f"Original number: {num}")
print(f"Reversed number : {reverseNum(num)}")
```

#### 6. String Num function finds the first number hidden in the string

```python
def strNum(s):

    sign = 1
    temp = ""

    for char in s:
        if char.isnumeric():
            temp += char
        elif char == "-" or char == "+" or char == " ":
            if char == "-":
                sign = -1
            if temp != "":
                return int(temp) * sign
    
    return int(temp) * sign

strung = "abc 123 abc"
print(f"Original string: {strung}")
print(f"Found number : {strNum(strung)}")
```
